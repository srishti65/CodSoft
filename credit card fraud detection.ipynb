```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
```

```python
x=data.drop('Class',axis=1)
y=data['Class'].ravel()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
len(x_train),len(x_test),len(y_test),len(y_train)
```

```python
Lreg=LogisticRegression(random_state=22)
Lreg.fit(x_train,y_train.ravel())
```

```python
y_predict=Lreg.predict(x_test).ravel()
y_predict
```

```python
data.Class
```

```python
Lreg.score(x_test,y_test)
```

```python
# CREATE CONFUSION MATRIX
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_predict)
score
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,y_predict).ravel()
print("True Negative",tn)
print("False Positive",fp)
print("False Negative",fn)
print("True Positive",tp)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

Here Sensitivity is 0.52 i.e,True Positive/TP+FN.
Here specificity is 1 i.e, True Negative/TN+FP

```python
ACCURACY=(tp+tn)/(tp+tn+fp+fn)
print("Accuracy:{:0.2f}".format(ACCURACY))
```

```python
PRECISION=(tp)/(tp+fp)
print("Precision{:0.2f}".format(PRECISION))
```

```python
RECALL=tp/(tp+fn)
print("Sensitivity{:0.2f}".format(RECALL))
```

```python
specificity=(tn)/(tn+fp)
print("specificity{:0.2f}".format(specificity))
```
